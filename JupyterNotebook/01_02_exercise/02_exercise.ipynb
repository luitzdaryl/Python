## 5.(Advanced) Nested/Dictionary Comprehension:

Dictionary comprehension works very similar to list comprehension, but instead a single value you are manipulating an entry of a `dict`. Quick example:

```python
values = range(5)

# for each number, find out if it is bigger than 3
# therefore, create a dictionary where the key is the number, and the entry's value is
# the answer (True or False)

{x: x>3 for x in values} # each entry consits of key:value
```
