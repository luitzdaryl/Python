### OOP: Constructor

The previous object wasn't able to do much, therefore lets build a `class` for a car and lets add some attributes and methods. Every car should has a name, therefore we can make sure this name is handed over to the car when we create it. We always create an object with the constructor of the `class`, which is the `__init__` function. This function is always called when we create an instance of the `class`. You probably have seen that the number of parameters in the `__init__` does not match the ones when we actually construct the object. It doesn't need to be `self`, but in Python it is widely spread and common practice.

In the following example one attribute is created over the reference the the object (`self`), and the other not. You will see that after creation, you cannot access the attribute (because it was never created).

### OOP:Inheritance

One major concept in OOP is inheritance. It allows us to quickly write a new `class` which inherits all attributes and methods from another class. Therefore you don't need to rewrite an entire `class` when you want to create a new `class`, based on a existing one. 

```python
class Mother():
    
    def __init__(self, fname, lname):
        self.fname = fname
        self.lname = lname
        
    def say_name(self):
        print(f"My name is {self.fname} {self.lname}.")
```

```python
mother = Mother("Sarah", "Connor")
mother.say_name()
```

```python
class Son(Mother): # the class "Son" inherits from "Mother", therefore has the same attributtes and methods
    pass
```

```python
son = Son("John", "Connor")
son.say_name()
```

A child `class` can also have new methods and attributes, or even change existing ones. If we change an inherited existing method, this is called __override__.

```python
# Parent class
class Animal:
    
    def __init__(self, name):
        self.name = name
        
    def say_name(self):
        print(f"My name is {self.name}.")
    
    def make_sound(self):
        print("")# print nothing, we dont know how all animals sound!     
        
class Cat(Animal):
        
    # the class Cat overrides the method "make_sound" from its parent class    
    def make_sound(self):
        print("Miau")

class Dog(Animal):
    
    def __init__(self, weight, likes_bone):
        super().__init__(weight)
        self.likes_bone = likes_bone
    
    # the class Dog overrides the method "make_sound" from its parent class  
    def make_sound(self):
        print("Woof")
        
    def dig_for_bone(self):
        if self.likes_bone:
            print("Found bone!")
        else:
            print("I dont like bones")
```

```python
cat = Cat("Tom")
dog = Dog("Dexter", True) # this dogs likes bone

# both objects can use the method "say_name", since they inherit from Animal
cat.say_name()
dog.say_name()
```

```python
# both can make noise, but since they override it, they make different noise
cat.make_sound()
dog.make_sound()
```

```python
# only the dog has the method dig for bone! cats dont do that
```

```python
dog.dig_for_bone()
```

```python
# cat will throw error
cat.dig_for_bone()
```

```python
# also if we create an animal object, it wont be able to dig for bones
a = Animal("Random")
a.dig_for_bone()
```

## File I/O

Usually, when we have a lot of data, we need to read it from some file. For that, python gives us two statements, which will let us manipulate files.
First one is `open`, which opens a file and returns a file object. The second argument tells us what kind of mode we want to enter:

__"r"__ - Read - Default value. Opens a file for reading, error if the file does not exist

__"a"__ - Append - Opens a file for appending, creates the file if it does not exist

__"w"__ - Write - Opens a file for writing, creates the file if it does not exist

__"x"__ - Create - Creates the specified file, returns an error if the file exist


The next one is the `with` statements, which handles the setup and closing of the file object. More under https://docs.python.org/2.5/whatsnew/pep-343.html.
Together, we can access file values. 

```python
with open('res/test_results.txt', 'r') as file:
    data = list()
    for line in file: # iterate over every line in .txt file
        
        line_clean = line[:-1] # remove the last character of that line, which is a cariage return symbol \r
        
        list_values_str = line_clean.split(",") # split the string at each "," into a list of strings
        
        for value_str in list_values_str: # iterate over all values
            value_float = float(value_str) # change value to float
            data.append(value_float) # add value to string
data
```

We can also do this a lot quicker with list comprehension.

```python
with open('res/test_results.txt', 'r') as file:
    data = list()
    for line in file:
        data.extend([float(x) for x in str.split(line[:-1], ',')]) # add all values in the line
data
```

Just as easy we can read _csv._ files (Comma Separated Values). They are just text files seperated, or delimited by a comma. CSV is a simple file format used to store tabular data, such as a spreadsheet or database. Files in the CSV format can be imported to and exported from programs that store data in tables, such as Microsoft Excel or OpenOffice Calc, and many others.

```python
with open('res/csv_example.csv', 'r') as f:
    D = list()
    for line in f: # split until -1 because we dont want charater return (\r)!
        D.append(tuple([float(x) if x.isnumeric() else x for x in str.split(line[:-1], ',')]))
print(D)
```

We can also create new files and store values we have calculated in an file with `with` and `open`. But we have to make sure all values are of type `str`.

```python
import random

values = [[random.randint(0,10) for _ in range(0,100)] for _ in range(10)]

with open("res/write_csv_example.csv", "w") as f:
    for line in values:
        # convert all values to string
        line = [str(x) for x in line]
        f.write(",".join(line)+"\n")
    
    f.write("\n")
```

## Lamda Expression and other more Built-ins
In Python, an anonymous function is a function that is defined without a name.

While normal functions are defined using the `def` keyword in Python, anonymous functions are defined using the `lambda` keyword.

Hence, anonymous functions are also called lambda functions.

The syntax for the lambda function is:

**lambda** **arguments**: **expression**

```python
power2 = lambda x: x ** 2
power2(5)
```

A `lambda` function can also take multiple inputs, but always only has one expression and therefore output.

```python
summarizer = lambda a, b, c : a + b + c
summarizer(5, 3, 8), summarizer(7, 4, 2) 
```

Technically, this function has no name. The `lambda` returns a function object which is assigned to the variable `power2` or `summarizer`. We can now call it as a normal function. The statement is nearly the same as:

```python
def power2(x):
    return x * 2
```

We use lambda functions when we require a nameless function for short expressions.

In Python, we generally use it as an argument to a higher-order function (a function that takes in other functions as arguments). Lambda functions are used along with built-in functions like `filter()`, `map()` etc. With it, we can get very similar behavior as with list comprehension.

```python
# Program to filter out only the even items from a list
my_list = [1, 5, 4, 6, 8, 11, 3, 12]

new_list = list(filter(lambda x: (x%2 == 0) , my_list)) # filter values out where the the values is not true

print(new_list)
```

```python
# Program to double each item in a list using map()
# similar as to manipulation part 
my_list = [1, 5, 4, 6, 8, 11, 3, 12]

new_list = list(map(lambda x: x * 2 , my_list))

new_list
```

`lambda` functions can also take multiple inputs.

```python
reduce = lambda x,y: x+y 

my_list1 = [1, 5, 4, 6, 8, 11, 3, 12]
my_list2 = [6, 3, 3, 9, 7, 26, 5, 1]

list(map(reduce, my_list1, my_list2)) # add all values together of both list
```

With `lambda`, we can get similar behavior as with list comprehension. My personal opinion ist, list comprehension is easier to read. It really depends on what you prefer. But as many times in programming, both are sufficient. It depends on what you like more and are more comfortable with.

There is one more aspect of `lambda` expressions we have not explored yet, therefore lets look at the type of the function.

## Packing and Unpacking

We can easily release and compress the content of python `list`, `tuple` and also `dict` using the packing operators.

- `*` for `list`, `tuple`
- `**` for `dict`

With packing and unpacking we can easily modify the input parameters of any function and therefore allow more flexibility when defining and handling parameters. 

```python
vector1 = [2,3,4]
```

```python
def norm(x1,x2,x3):
    
    value = (x1**2 + x2**2 + x3**2)**0.5
    
    return value
```

```python
norm(vector1) # this will throw error, because the paramters of the function dont match up
```

We can unpack the values of the `list` and hand over each element to the function respecting their order in the list.

```python
norm(*vector1) # unpack all values in the list
```

The unpacked amount needs to match the number of parameters of the function.

```python
vector2 = [1,2,3,4]
norm(*vector2) # will throw an error! 
```

We can also use packing to allow any amount of arguments for our function. Therefore we simply pack all parameters in our function header together in one tuple. Common practice is to use the parameter __\*args__ to inform that an unlimited number of parameters are put together into one tuple.

```python
def make_sentence_from_words(*args):
    
    print(type(args)) # we see *args is just an tuple
    
    sentence = "" # empty string
    
    for word in args:
        sentence += " "+ word # concat with whitespace 
    
    sentence += "."
    return sentence
```

```python
make_sentence_from_words("I","love","data","science","with","python")
```

The built-in `zip` function also uses packing to zip any number of lists together.

```python
list1 = list(range(5))
list2 = list(range(8, 8+5))
list3 = list(range(3, 3+5))
list4 = list(range(11, 11+5))

list(zip(list1,list2)), list(zip(list1,list3,list2))
```

```python
# we can also put all values in a tuple, and them unpack them again
all_values = (list1, list2, list3, list4)
list(zip(*all_values)) # unpack and then pack again, maybe a little redundant...
```

## Iterators and Generators

Iterators are objects that traverse through a given sequence. An iterator can be iterated upon, meaning that you can go from one element to the next. Therefore, you can not jump back, since each element in the iterator only knows who its next element is. 

Classes like `list`, `dict`, `tuple` and `set` are __iterable__ objects. They are __iterable__ containers, from which you can get an iterator from with `iter(iterable)`. 

The difference is, the containers actually save and structure their elements a certain way, while an iterator only knows what the next element is. 

We can use the builtin function `iter()`, to get the iterator from a container.

With `next()` we can get the next element of the iterator.

The `for` loop actually calls upon the iterator of an container, to iterate over all elements in the container.

In the following example, we can see whats actually going on in a `for` loop.

```python
# value can be anything which can be iterate 
values = ['value1','value2', 'value3']

# with for loop:
for value in values:
    print(value)
```

```python
# we can do exact same thing with the iterator,
# which we can get with iter()

values_iter = iter(values)

for value in values_iter:
    print(value)
```

The iterator and the `list` might seem equal, but they are different objects:

![](img/list_vs_iterators.png)

```python
# whats actually going on (behind the scene)
iterator = iter(values) # get the iterator
while True: 
    try: # need to 
        # Iterate by calling next 
        item = next(iterator) 
        print(item) 
    except StopIteration: 
        # exception will happen when iteration will over
        print("Loop is over") # this will not be printed usually!
        break
```

```python
iterator.__next__() # will throw error, because iterator is at the end!
```

Therfore, an iterator only knows the next element,as shown in the following picture:

`enumerate` and `zip` are iterators, while the builtin functions of the same name generate the iterator object. We can call the builtin `next` function.

Each iterator has to implement the two `__iter__()` and `__next__()` (Iterator Protocol).

We can also create our own iterator, by implementing the Iterator Protocol in a `class`:

```python
class HahaIterator:
    
    def __init__(self, start=0, end=10):
        self.count = start
        self.end = end
    
    def __iter__(self):
        return self
    
    def __next__(self):
        
        element = "ha"*self.count # mutiply our laugh by iteration
        self.count = self.count + 1 # increase count
        
        if self.count > self.end: # if we dont consider this, it will go forever!
            raise StopIteration
        
        return element
```

```python
haha_iter = HahaIterator()

for x in haha_iter:
    print(x)
```

## Generators

With Generators we can easily create our own iterators, without having to write an entire `class` with the Iterator Protocol. There are two ways: __generator expression__ and __generator function__.

### Generator Expression

A generator expression is basically list comprehension for iterators. Therefore we can use the same exact syntax, only we use the `()` insted of `[]`.

```python
gen = (x for x in range(9)) # exactly the same...
gen
```

```python
next(gen) 
```

```python
gen = (x for x in range(9)) 

for x in gen:
    print(x)
```

### Generator Function

For more sophisticated generators, we can also create a function, which returns a generator. For that we need to use the `yield` statement, and __NO__ `return`!!

```python
def gen_func():
    # genarator function with only one value
    for x in range(9):
        yield 
        {
        "A":x
        }
```

```python
x = gen_func()
```

```python
next(x)
```

```python
next(x)
```

In the following we will go over a quick example, where we want to use a regex code to see if the "code" matches another string. Therefore we need to write the regex code as a Python string, and then compile it using the regex library.

In our example, we want to match the string _"hello"_. Therefore we first need to compile our regex expression. It is a own programming language after all.

```python
p = re.compile("hello") # this returns a re compile object
p
```

When we match our regex to another string, we can use the `match()` method.

```python
result = p.match("hello")
result
```

The `re.Match` object tells us about the result, and we can also use some methods on it to get these reseults.

```python
result.group() # return the string matched by the RE
```

```python
result.start() # Return the starting position of the match
```

```python
result.end()#  Return the ending position of the match
```

```python
result.span() #return the tuple with start and end
```

We can also use other methods on the compiled regex object to search a pattern in a string.

```python
result_search = p.search("this is a string, lets look if hello is in here.")
result_search.span() # get start and end posistion
```

If there is not a match or the search failed, the `match()` and `search()` will return `None`. 

```python
search_result2 = p.search("That word is not in here")

if search_result2:
    print(search_result2.group())
else:
    print("no result!")
```

### Metacharacters

Until now, we have not really done anything special, what we couldn't also do with `str` methods. To code patterns in regex, we need to use metacharacters. When we write the regex code/pattern, most characters are used as they are. There are a few exceptions, which are the metacharacters. Here are the most basic ones we will discuss in this course:

- `[]`: match a set and ranges of characters
- `*`: match a repetition of preceading characters (including 0 repetitions)
- `+`: match a repetition of preceading characters (at least 1 repetition)
- `|`: match either characters A`|`B  
- `{m,n}` : match a number of repetitions of the preceading characters
- `()`: captures a group of RE 

```python
re.search("[a-z]+@[a-z]+.[a-z]+","hdhdhddd@sgmail.com")
```

Lets improve this, because my email for example wouldn't not get a result :( 

```python
re.search("([a-z|.])+@[a-z]+.[a-z]+","harris.e.jason@sgmail.com")
```

Repeating operators such as `*`, `+` and `{}` are greedy, which means they will always try to find the longest string possible.

```python
re.search("[a-z|.]+@[a-z]+.[a-z]+","harris.e.jason@sgmail.thisiswaytoolong")
```

Maybe we could add an allowed `set` of values to our regex? Since our regex is just a string, we can also code the regex with python.

```python
allowed_email_endings = {"com","de","net","org"} #definetly there are more!

x = "|".join(allowed_email_endings) # join them together with or
x = "((com)|(de)|(org)|(net))"
```

```python
re.search(f"[a-z|.]+@[a-z]+.{x}","harris.e.jason@sgmail.thisiswaytoolong")
#result is None # result is None, which is good! The string above is clearly not a email
```

```python
re.search(f"[a-z|.]+@[a-z]+.{x}","harris.e.jason@gmail.net")
```

### Groups

Many times we allow a certain range or set of characters. Regex therefore has many groups already defined for us: The complement of a group A is the group, which contains everything which is not in group A.


- `\d`: group for digits; this is equivalent to the class `[0-9]`.
- `\D`: group for non-digit characters; complement of `\d`
- `\s` : group of whitespace characters (cariage return etc) ; this is equivalent to the class `[ \t\n\r\f\v]`.
- `\S`: complement of `\s`
- `\w`: group of alphanumeric characters; this is equivalent to the class `[a-zA-Z0-9_]`.
- `\W`: complement of `\w`

Note: we can also build our own compliments with `[^A]`.  


```python
re.search(f"[\w|.]+@[\w]+.{x}",big_string)
```
